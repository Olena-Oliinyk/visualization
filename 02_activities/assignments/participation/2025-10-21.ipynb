```python
#load our libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#make some sample data
np.random.seed(613)
x = np.arange(50)
y = np.random.randint(0,100,50)

#make a line plot
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)

#adding axis labels and titles
ax.set_title('Total growth over time')
ax.set_ylabel('Total growth')
ax.set_xlabel('Years since start')
```

```python
#make a dictionary with the properties we want for our labels
font1 = {'family' : 'sans-serif', 'color': 'blue', 'size':20 }
font2 = {'family' : 'monospace', 'color': 'green', 'size':14 }

#dd fontdict arguments to our previous lineplot
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)

ax.set_title('Total growth over time', fontdict = font1, loc = 'left') #move our lables with the loc argument
ax.set_ylabel('Total growth ', fontdict = font2)
ax.set_xlabel('Years since start', fontdict = font2)
```

```python
#modify data points
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y, 
marker = '*', 
markersize = 12, 
color = 'indigo',
linestyle = '--',
linewidth = 2,
markeredgecolor = '#fa9359',
markerfacecolor = '#000000')

#add grid lines
ax.grid(axis = 'y')
```
